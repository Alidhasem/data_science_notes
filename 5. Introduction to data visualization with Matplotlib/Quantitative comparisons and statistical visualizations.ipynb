```
# Call the ax.bar method to plot the "Gold" column as a function of the country
import pandas as pd 
import matplotlib.pyplot as plt

medals = pd.read_csv('/work/data_science_notes/5. Introduction to data visualization with Matplotlib/data/medals_by_country_2016.csv', index_col=0)
fig, ax = plt.subplots()
colors = ['b', 'violet', 'g', 'r', 'c', 'm', 'y', 'b', 'orange', 'deeppink']
plt.bar(medals.index, medals.Gold, color=colors)

# Use the ax.set_xticklabels to set the x-axis tick labels to be the country names
# In the call to ax.set_xticklabels rotate the x-axis tick labels by 90 degrees by using the rotation key-word argument
ax.set_xticklabels(medals.index, rotation=90)

# Set the y-axis label to "Number of medals"
ax.set_ylabel("Number of gold medals")
plt.show()
```

```
import seaborn as sns

fig, ax = plt.subplots()
plt.pie(medals.Gold, labels=medals.index)
plt.show()
```

```
# Create a scatter plot with "co2" on the x-axis and "relative_temp" on the y-axis
climate_change =  pd.read_csv('/work/data_science_notes/5. Introduction to data visualization with Matplotlib/data/climate_change.csv')
fig, ax = plt.subplots()

sns.set()
ax.scatter(climate_change.co2, climate_change.relative_temp, c=climate_change.index)

# Set the x-axis label to "CO2 (ppm)"
# Set the y-axis label to "Relative temperature (C)"
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (C)')
```
