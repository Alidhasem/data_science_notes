# Matplotlib

## Basic plots with Matplotlib

### Data visualization

Data visualization is one of the fundamental steps of data analysis. It is useful to find insights in our data and also for sharing our results.

### Matplotlib

Matplotlib is the fundamental library for data visualization with Python. Usually you are going to need the `pyplot` module from the package.

```python
import matplotlib.pyplot as plt
```

### Line plot

Line plot is the most basic kind of plot you can build with Matplotlib. It is useful, for example, when you have time as your x-axis variable.

```python
import pandas as pd
world_pop = pd.read_csv('./data/world_population.csv')
year = world_pop['year']
population = world_pop['population']
```

```python
# Use plt.plot() to build a line plot, year should be mapped on the horizontal axis, pop on the vertical axis
plt.plot(year, population)
plt.show()
```

### Scatter plot

A scatter plot is a good option when you're trying to assess if there's a correlation between two variables.

```python
gdp = pd.read_csv('./data/gapminder.csv', index_col=0)
pop = gdp['population']
life_exp = gdp['life_exp']
gdp_cap = gdp['gdp_cap']
cont = gdp['cont']
```

```python
# Build a scatter plot, where gdp_cap is mapped on the horizontal axis, and life_exp is mapped on the vertical axis
plt.scatter(gdp_cap, life_exp)
plt.show()
```

```python
# A correlation will become clear when you display the GDP per capita on a logarithmic scale. Add the line 
plt.scatter(gdp_cap, life_exp)
plt.xscale('log')
plt.show()
```

## Histogram

### Histograms

A histogram shows the distribution of a variable across ranges of its possible values. It is useful for taking a quick look at what values are more common for a variable in a dataset.

```python
# Use plt.hist() to create a histogram of the values in life_exp 
# Do not specify the number of bins; Python will set the number of bins to 10 by default for you
plt.hist(life_exp)
plt.show()
```

### The `bins` argument

It is very importatn to specifiy the number of bins your histogram will have. Too few bins could result in an oversimplification of reality and you could miss important details. Too many bins could overcomplicate your plot and you might not be able to get any insights.

```python
# Build a histogram of life_exp, with 5 bins
plt.hist(life_exp, bins=5)
plt.show()
```

```python
# Build another histogram of life_exp, this time with 20 bins
plt.hist(life_exp, 20)
plt.show()
```

## Customization

### Customization guidelines

Your plot must display the information in a useful way. This is, show the data in a way that it shares a message. To accomplish this, you need to cover the next points in your plot customization.

* Context of your data (What are you plotting?). Labels and titles are useful to cover this point.  
* Message of your plot. In order to effectively transmit your message, you will need to choose the type of plot to use, its axis scales and ticks, colors and more.

```python
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']
dict = {
    'Asia':'red',
    'Europe':'green',
    'Africa':'blue',
    'Americas':'yellow',
    'Oceania':'black'
}

# Import the numpy package as np
import numpy as np

# Use np.array() to create a numpy array from the list pop 
# Call this NumPy array np_pop
np_pop = np.array(pop)

# Double the values in np_pop setting the value of np_pop equal to np_pop * 2
# Multiply for (10**-6) to get numbers with only one whole digit 
np_pop = np_pop*2*(10**-6)

# Use dict to build a list containing the corresponding color for each color in cont
col = []
for i in cont:
    if i == 'Asia':
        col.append('red')
    elif i == 'Europe':
        col.append('green')
    elif i == 'Africa':
        col.append('blue')
    elif i == 'Americas':
        col.append('yellow')
    else:
        col.append('black')


# Build a scatter plot, where gdp_cap is mapped on the horizontal axis, and life_exp is mapped on the vertical axis
# Set the s argument inside plt.scatter() to be np_pop
# Add c = col to color the bubbles
#  Change the opacity of the bubbles by setting the alpha argument to 0.8
plt.scatter(gdp_cap, life_exp, s=np_pop, c=col, alpha=0.8)

# Use a logarithmic scale for the x-axis
plt.xscale('log')

# Use xlab and ylab to set the label of the x- and y-axis
plt.xlabel(xlab)
plt.ylabel(ylab)

# Use the string title to add a title to the plot
plt.title(title)

# Use tick_val and tick_lab as inputs to the xticks() function to make the the plot more readable
plt.xticks(tick_val, tick_lab)

# Add labels for China and India
plt.text(1550, 71, 'India')
plt.text(5700, 80, 'China')

# Add gridlines to the plot
plt.grid(True)

# Show the plot
plt.show()
```
